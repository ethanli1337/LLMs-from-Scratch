```python
import numpy as np
import torch 
import torch.nn as nn 
import matplotlib.pyplot as plt 

```

```python
z=[1,2,3]
num=np.exp(z)
den=np.sum(np.exp(z))
sigma=num/den

print(sigma)
print(np.sum(sigma))
```

```python
z=np.random.randint(-5,high=15,size=25)
print(z)

num=np.exp(z)
den=np.sum(num)
sigma=num/den
plt.yscale('log')
plt.plot(z,sigma,'ko')
plt.show()
```

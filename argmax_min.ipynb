```python
import numpy as np 
import torch 
import torch.nn as nn 
```

```python
v=np.array([1,40,2,-3])
minval=np.min(v)
maxval=np.max(v)

print(minval, maxval)

minidx=np.argmin(v)
maxidx=np.argmax(v)

print(minidx, maxidx)
```

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
p=0.25 
q=1-p 

H=-(p*np.log(p))
```

```python
x=[0.25,0.75]

H=0
for p in x:
    H-=p*np.log(p)
print(H)
```

```python
#Binary entropy
H=-(p*np.log(p)+(1-p)*np.log(1-p))
print(H)
```

```python
p=[1,0]
q=[0.25,0.75]

H=0
for i in range(len(p)):
    H-=p[i]*np.log(q[i])
print(H)
```

```python
import torch 
import torch.nn.functional as F 

q_tensor=torch.Tensor(q)
p_tensor=torch.Tensor(p)


F.binary_cross_entropy(q_tensor,p_tensor)
```
